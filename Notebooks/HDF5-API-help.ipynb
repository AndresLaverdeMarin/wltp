# HDF5 API:
list usefull HDF5 methods

```python
import sys
import wltp
from pathlib import Path

## Add tests/ into `sys.path` to import `vehdb` module.
#
proj_dir = str(Path(wltp.__file__).parents[1] / "tests")
if proj_dir not in sys.path:
    sys.path.insert(0, proj_dir)

import vehdb
```

```python
def print_h5API(displaywidth=160):
    from columnize import columnize
    from pandas import HDFStore

    with vehdb.openh5('WltpCars.h5') as h5db:
        print('HDFStore:', 
              columnize([f for f in dir(h5db) if not f.startswith('__')], displaywidth=displaywidth),
              '%s (accessed by `<HDFStore-instance>._handle`):' % type(h5db._handle), 
              columnize([f for f in dir(h5db._handle) if not f.startswith('__')], displaywidth=displaywidth),
              '%s:' % type(h5db.root), 
              columnize([f for f in dir(h5db.root) if not f.startswith('__')], displaywidth=displaywidth),
              '%s:' % type(h5db.root._v_attrs), 
              columnize([f for f in dir(h5db.root._v_attrs) if not f.startswith('__')], displaywidth=displaywidth),
             sep='\n')
print_h5API()
```
